```python
# importing neccessary libraries
import pandas as pd
import numpy as np
import matplotlib
import seaborn as sns
```

```python
# creating variable for dataset display and testing it
file = 'NBA_Player_Stats_2.csv'
df = pd.read_csv(file)
print(df.head())
```

```python
# isolating MVPs from rest of players in dataset
df1 = df[df['MVP']  == True]
df1
```

```python
# adding new column so that the year will be seen as an integer and not a text, which makes it much easier to work with and handle
df1['season-year'] = df1['Season'].str[:4]
df1['season-year'] = pd.to_numeric(df1['season-year'])
df1
```

```python
# isolating specific time span for MVP dataset
df2 = df1[(df1['season-year'] >= 2008) & (df1['season-year'] <= 2018)]
df2
```

```python
# creating bar graoh for points per game MVP comparison
ax = df2.plot.bar(x = 'Player', y = 'PTS')
```

```python
# creating bar graoh for assists per game MVP comparison
ax = df2.plot.bar(x = 'Player', y = 'AST')
```

```python
# creating bar graoh for rebounds per game MVP comparison
ax = df2.plot.bar(x = 'Player', y = 'TRB')
```
